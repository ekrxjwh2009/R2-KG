```python
import json
def load_from_json(file_path):
    """JSON 파일에서 데이터를 로드"""
    with open(file_path, 'r', encoding='utf-8') as f:
        return json.load(f)

claim_path = f"./results_final/Paraphrase/gpt-4o-mini/paraphrased_claims.json"
paraphrased_claims = load_from_json(claim_path)
```
