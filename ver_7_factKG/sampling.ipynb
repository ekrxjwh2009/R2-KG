```python
import json
```

```python
questions_dict = {}
entity_set_dict = {}
label_set_dict = {}
types_dict ={}
existence, num_1, multihop, multiclaim = [],[],[],[]
```

```python
f= open("/home/smjo/share_code/factkg/data/extracted_test_set.jsonl")
for line in f:
    if not line:
        continue
    q = json.loads(line)

    q_type = q["types"]
    if "existence" in q_type:
        existence.append(q["question_id"])
    elif "multi claim" in q_type:
        multiclaim.append(q["question_id"])
    elif "multi hop" in q_type:
        multihop.append(q["question_id"])
    elif "num1" in q_type:
        num_1.append(q['question_id'])
    else:
        continue
```
